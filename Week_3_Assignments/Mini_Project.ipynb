```python
#Count the null values 
df.isnull().sum()
```

```python
#drop the missing values
missing_values= df.dropna()
missing_values
```

```python
dropped_rows = df[df.isna().any(axis=1)]
print(dropped_rows)
```

##  Group By (count of Manufacturer by city)

```python
city_group = df.groupby("City")["Manufacturer"].count().sort_values(ascending=True)
print(city_group.head())
```

## Sorting (sort by Ship date descending)

```python
sorted_df = df.sort_values(by='Ship Date',ascending=False)
print(sorted_df.head())
```

## filtering(profit>20)

```python
# Clean the Profit column (remove $ and convert to integer)
df["Profit"] = df["Profit"].replace('[\$,]', '', regex=True).astype(int)
# Find all rows where Profit is 20
result = df[df["Profit"] >=20]
result.head()
```

## Aggregations (sum of Quantity by category)

```python
_sum_ = df.groupby('Category')['Quantity'].sum()
print(_sum_)
```

```python
# Multiple aggregation 
revenue_stats = df.groupby('City')['Quantity'].agg(['count', 'mean', 'sum', 'max', 'min'])
print(revenue_stats)
```

## Visualization

```python
subcat_profit = df.groupby('Sub-Category')['Profit'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(x=subcat_profit.values, y=subcat_profit.index, palette='coolwarm')
plt.title('Average Profit by Sub-Category')
plt.xlabel('Average Profit')
plt.ylabel('Sub-Category')
plt.tight_layout()
plt.show()
```

```python
shipmode_counts = df['Ship Mode'].value_counts()

plt.figure(figsize=(7, 5))
sns.barplot(x=shipmode_counts.index, y=shipmode_counts.values, palette='Set2')
plt.title('Number of Orders by Ship Mode')
plt.ylabel('Number of Orders')
plt.xlabel('Ship Mode')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.violinplot(data=df, x='Segment', y='Profit', palette='muted')
plt.title('Profit Distribution by Customer Segment')
plt.xlabel('Segment')
plt.ylabel('Profit')
plt.tight_layout()
plt.show()
```
